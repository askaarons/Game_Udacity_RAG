# Udaplay Part 1 - RAG Pipeline

Load game JSON files, process documents, embed into ChromaDB, and run semantic search.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.rag.data_processor import load_game_records, to_documents
from src.rag.vector_store_manager import VectorStoreManager
```

```python
records = load_game_records(PROJECT_ROOT / 'data' / 'raw')
documents = to_documents(records)
vector_store = VectorStoreManager(persist_directory=PROJECT_ROOT / 'chroma_db', collection_name='games')
inserted = vector_store.add_documents(documents)

print('Loaded records:', len(records))
print('Inserted/updated docs:', inserted)
print('Collection count:', vector_store.count())
```

```python
queries = [
    'Which game is a farming simulation?',
    'What game is developed by FromSoftware?',
    'Tell me a Nintendo open world action-adventure game'
]

for query in queries:
    print('=' * 80)
    print('Query:', query)
    for rank, item in enumerate(vector_store.semantic_search(query, top_k=3), start=1):
        print(rank, item.get('metadata', {}).get('title'), item.get('similarity'))
```

## Outcome
Semantic retrieval is working over the persistent local Chroma collection.
