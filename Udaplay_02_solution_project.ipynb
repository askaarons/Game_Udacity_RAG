# Udaplay Part 2 - Agent Implementation

Stateful agent with `retrieve_game`, `evaluate_retrieval`, and `game_web_search` tools.

```python
from dataclasses import asdict
from pathlib import Path
import json
import sys

PROJECT_ROOT = Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.agent.reporting import build_agent_report
from src.agent.state_machine import GameRAGAgent
from src.rag.vector_store_manager import VectorStoreManager
```

```python
vector_store = VectorStoreManager(persist_directory=PROJECT_ROOT / 'chroma_db', collection_name='games')
agent = GameRAGAgent(vector_store=vector_store)
print('Loaded vector store count:', vector_store.count())
```

```python
example_queries = [
    'When was Elden Ring released and on which platforms?',
    'Who published Stardew Valley?',
    'What can you tell me about GTA VI release window?'
]

for i, query in enumerate(example_queries, start=1):
    turn = agent.run(query)
    report = build_agent_report(asdict(turn))
    print('=' * 100)
    print(f'Example {i}: {query}')
    print(json.dumps(report, indent=2))
```
